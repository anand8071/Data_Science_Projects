```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
from scipy.stats import chi2_contingency
```

```python
BuyerRatio= pd.read_csv('/content/BuyerRatio.csv')
BuyerRatio
```

```python
#  Make Two Dimensional Array
Buyerratio= np.array([[50,142,131,70],[435,1523,1356,750]])
Buyerratio
```

##Apply Chi-Square table contigency table to convert Observed Value into Expected Value


```python
stats, p, dof, exp = stats.chi2_contingency(Buyerratio)
print(stats,"\n", p, "\n",dof, "\n",exp)
```

#P value of Chi-Square test == 0.66 > 
##Hence we accept null hypothesis i.e 
##Male and Female buyers are similar across the region
